```python
import os
import sys
import numpy as np
sys.path.append('..')
import requests
import json
from pytube import YouTube
import youtube_dl
import cv2
from os import listdir
import shutil
```

```python
with open('../data/activity_net.v1-2.min.json') as data_file:    
    data_json = json.load(data_file)
    
with open('../data/videos/training/training_meta_data.json', 'r') as data_file:
    metadata_json = json.load(data_file)
```

```python
path = '../data/videos/training'
classes_training = 0
classes_amount = 101
videos_metadata = {}
resize=(224, 224)
for file_name in os.listdir(path):
    video_id = file_name.split('.')[0]
    if video_id in data_json['database']:
        cap = cv2.VideoCapture(path + '/' + file_name)
        fps = cap.get(cv2.CAP_PROP_FPS)
        frame_count = 0
        frames = []
        labels = []
        print('annotations')
        for item in data_json['database'][video_id]['annotations']:
            print(item['segment'][0], item['segment'][1], item['label'])
        print('-------------')
        try:
            while True:
                ret, frame = cap.read()
                if not ret:
                    break
                frame_count += 1
                if frame_count % 6 == 0:
                    label = np.zeros(classes_amount)
                    label[0] = 1
                    seconds = frame_count / fps
                    for item in data_json['database'][video_id]['annotations']:
                        if seconds < item['segment'][1] and seconds > item['segment'][0]:
                            label[metadata_json[item['label']]['idx']] = 1
                            label[0] = 0
                            break
                    frame = cv2.resize(frame, resize)
                    frame = frame[:, :, [2, 1, 0]]
                    frames.append(frame)
                    labels.append(label)
                
        finally:
            cap.release()
        print(len(frames))
        break
```
